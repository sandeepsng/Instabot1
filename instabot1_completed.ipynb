```python
def food_related_usernames():
    
#     locatig search box and clicking on it to activate
    search_box=driver.find_element_by_xpath("//div[@class='pbgfb Di7vw ']")
    search_box.click()
    
#     waiting for page to load
    time.sleep(2)

#     locatig search box input
    search__input=driver.find_element_by_xpath("//input[@class='XTCLo x3qfX focus-visible']")

#     sending search query i.e "food" to the search box
    search__input.send_keys("food")
    
#     waiting for dialog box with username and hashtags list to arrive
    time.sleep(4)

#     locating the dialog box
    drop_down=driver.find_element_by_xpath("//div[@class='fuqBx']")

#     converting the data to text to extract list of usernames
    data=drop_down.get_attribute("innerHTML")

    
    
        ############################# DATA CLEANING STARTS HERE ###################################
                            
#     for seperating each individual elment of grid/list
    l=data.split("<a")

#     creating a new list discarding all the hash tags
    new_l=[]
    for e in l:
        if "explore/tags" in e or len(e)==0:     #for eliminating hashtags and empty string 
#                                                 created due to split(one empty at last)) 
            None
        else:
            new_l.append(e)

    
#     this function discard all uneccesary text such as tags/ classes etc
    def clean(each):
        start_ind=each.index("Ap253")
        valuable=each[start_ind+7:]
        end_ind=valuable.index("<")
        absolute=valuable[:end_ind]
        return absolute


    list_of_usernames=[]
    
#     iterating over each element created by split at passing it to cleaning function
    for j in new_l:
        single_user=clean(j)
        list_of_usernames.append(single_user)
        
    return list_of_usernames     #returning a list of clean usernames
```

```python
import time
def follow_unfollow_message(username,action):
    
#     locatig search box and clicking on it to activate
    search_box=driver.find_element_by_xpath("//div[@class='pbgfb Di7vw ']")
    search_box.click()
    
#     waiting for page to load
    time.sleep(2)

#     locatig search box input
    search__input=driver.find_element_by_xpath("//input[@class='XTCLo x3qfX focus-visible']")

#     sending search query i.e "So delhi " to the search box
    search__input.send_keys(username)
    
#     waiting for dialog box with username and hashtags list to arrive
    time.sleep(4)
    
    
    #     creating profile specific link to profile href usinga general format
    href_link="//a[@href='/"+username+"/']"

#      opening desired profile 
    desired_profile=driver.find_element_by_xpath(href_link)
    desired_profile.click()
    

    time.sleep(2)
    
    
    
                               ##################################
   ############################################################################################ 
                               ##################################
                        #Decision block acording to the passed action
    
    
    if action=="follow":    #if Passed action is follow
        try:
            #finding follow button and clicking on it
            follow_button=driver.find_element_by_xpath("//button[@class='_5f5mN       jIbKX  _6VtSN     yZn4P   ']")
            follow_button.click()
        
        
        except:
            #if follow button is not found we will recieve a message "Already Foloowing"
            print("Already Following")
    
    
    elif action=="unfollow":    #if passed ation is unfollow
        try:
            
            #finding unfollow button  and clicking on it
            unfollow_button=driver.find_element_by_xpath("//button[@class='_5f5mN    -fzfL     _6VtSN     yZn4P   ']")
            unfollow_button.click()
            
            time.sleep(1)
            
            #clicking on unfollow button to confirmation message
            confirm_unfollow=driver.find_element_by_xpath("//button[@class='aOOlW -Cab_   ']")
            confirm_unfollow.click()
            
            
        except:
            #message will pop up if already unfollowed/ not currently following
            print("Already Unfollowed / Not following currently")
        
```

```python
import time
def like_unlike(username,action,num_posts):
    
#     locatig search box and clicking on it to activate
    search_box=driver.find_element_by_xpath("//div[@class='pbgfb Di7vw ']")
    search_box.click()
    
#     waiting for page to load
    time.sleep(2)

#     locatig search box input
    search__input=driver.find_element_by_xpath("//input[@class='XTCLo x3qfX focus-visible']")

#     sending search query i.e "So delhi " to the search box
    search__input.send_keys(username)
    
#     waiting for dialog box with username and hashtags list to arrive
    time.sleep(4)
    
    #     creating profile specific link to profile href usinga general format
    href_link="//a[@href='/"+username+"/']"

#      opening desired profile 
    desired_profile=driver.find_element_by_xpath(href_link)
    desired_profile.click()
    

    
    time.sleep(3)
    
    first_post= driver.find_element_by_xpath("//div[@class='v1Nh3 kIKUG  _bz0w']") 
    first_post.click()
    
    time.sleep(3)
     
#     finding like button and clicking on it
    like_button=driver.find_element_by_xpath("//span[@class='fr66n']") 
    data=(like_button.get_attribute("innerHTML")).split('"')
    like_status=data[5]
    if like_status!=action:
        print("You have already", str(action)+"d" ,"this post")
    else:
        like_button.click()
    
    time.sleep(1)
    for count in range(num_posts-1):
        next_button=driver.find_element_by_xpath("//a[@class=' _65Bje  coreSpriteRightPaginationArrow']") 
        next_button.click()
        time.sleep(4)
        like_button=driver.find_element_by_xpath("//span[contains(@class,'fr66n')]") 
        data=(like_button.get_attribute("innerHTML")).split('"')
        like_status=data[5]
        if like_status!=action:
            print("You have already", str(action)+"d" ,"this post")
        else:
            like_button.click()
        
```

## 6.1 Extract the usernames of the first 500 followers of ‘foodtalkindia’ and ‘sodelhi’

```python
import time
def extract_followers(username,number_of_followers_to_be_extracted):
    
#     locatig search box and clicking on it to activate
    search_box=driver.find_element_by_xpath("//div[@class='pbgfb Di7vw ']")
    search_box.click()
    
#     waiting for page to load
    time.sleep(2)

#     locatig search box input
    search__input=driver.find_element_by_xpath("//input[@class='XTCLo x3qfX focus-visible']")

#     sending search query i.e username to the search box
    search__input.send_keys(username)
    
#     waiting for dialog box with username and hashtags list to arrive
    time.sleep(2)
    
    
#     creating profile specific link to profile href usinga general format
    href_link="//a[@href='/"+username+"/']"

#      opening desired profile i.e "dilsefoodie"
    desired_profile=driver.find_element_by_xpath(href_link)
    desired_profile.click()
    
    time.sleep(2)
    
#     finding "followers" button and clicking on it
    follower_button=driver.find_element_by_xpath("//a[@class='-nal3 ']")
    follower_button.click()
    
    time.sleep(2)
    # one scroll gives additional 12 followers
    # so to get "N" followers we have to scroll "N//12" times
    limit=(number_of_followers_to_be_extracted//12)
    time.sleep(2)
    for countdown in range(limit):
        element = driver.find_element_by_xpath("//div[@class='oMwYe']")
        driver.execute_script("return arguments[0].scrollIntoView();", element)
#         print(countdown)
        time.sleep(3)
        
    # for extracting text from html
    individual_follower=driver.find_element_by_xpath("//div[@class='PZuss']").text
    
    
                     ############################################
#############################################################################################
#                           DATA CLEANING STARTS HERE

#     creating list
    follower_username=individual_follower.split("\nFollow")
    
    only_usernames=[]
    i_j=0
    for fff in follower_username:
        if fff=="":  #for tackling empty strings created by split
            continue
        if i_j==0:   # for tacking first element as it will be split into 2 sep. elements 
                        #while others will be split into 3 elements
            lp=fff.split("\n")
            only_usernames.append(lp[0])
            i_j+=1
        else:
            lp=fff.split("\n")
            only_usernames.append(lp[1])
    
    
    #as we only need first 500 followers
    only_usernames=only_usernames[:500]
    
    return only_usernames
        
```

```python
# RUN THIS CELL TO GET 500 FOLLOWERS OF SODELHI

followers_of_sodelhi=extract_followers("dilsefoodie",500)

# printing the result
for every_del in followers_of_sodelhi:
    print(every_del)
```

## 6.2. Now print all the followers of “foodtalkindia” that you are following but those who don’t follow you.


```python
# driver.back()
import time
def extract_followers(username):
    
#     locatig search box and clicking on it to activate
    search_box=driver.find_element_by_xpath("//div[@class='pbgfb Di7vw ']")
    search_box.click()
    
#     waiting for page to load
    time.sleep(2)

#     locatig search box input
    search__input=driver.find_element_by_xpath("//input[@class='XTCLo x3qfX focus-visible']")

#     sending search query i.e username to the search box
    search__input.send_keys(username)
    
#     waiting for dialog box with username and hashtags list to arrive
    time.sleep(3)
#     creating profile specific link to profile href usinga general format
    href_link="//a[@href='/"+username+"/']"

#      opening desired profile i.e "dilsefoodie"
    desired_profile=driver.find_element_by_xpath(href_link)
    desired_profile.click()
    
    time.sleep(2)
    
#     finding "followers" button and clicking on it
    follower_button=driver.find_element_by_xpath("//a[@class='-nal3 ']")
    follower_button.click()
    
    time.sleep(2)
    
    
#     creating a list of followers of this page
    list_followers_this_page=interact_with_dialog_and_extract()
    
    followers_this_page_set=set(list_followers_this_page)
#     print("page followers",followers_this_page_set)
#     print("page followers len",len(followers_this_page_set))
    
    
#     getting to home feed
    driver.back()
    
    time.sleep(3)
    
#     searching butto to open my profile and clicking on it
    my_profile=driver.find_element_by_xpath("//a[@class='_2dbep qNELH kIKUG']")
    my_profile.click()
    
    time.sleep(3)
    
    #     finding "followers" button and clicking on it
    follower_button=driver.find_element_by_xpath("//a[@class='-nal3 ']")
    follower_button.click()
    
    time.sleep(3)
    
    
#     creating a list of my followers
    list_myfollowers=interact_with_dialog_and_extract()

    myfollowers_set=set(list_myfollowers)
#     print("my followers",myfollowers_set)
#     print("my followers len",len(myfollowers_set))
    
#     closing dialog box (i.e followers list box)
    driver.back()
    
    time.sleep(3)
#     finding "following" button on my profile page and clicking on it
    my_following_button=driver.find_element_by_xpath("//a[contains(@href,'following')]")
    my_following_button.click()
    
    time.sleep(3)
    
    
#     creating a list of my followers
    my_following=interact_with_dialog_and_extract()
    
    my_following_set=set(my_following)
#     print("my_following_set",my_following_set)
        
#     mathematics to filter required people using set theory
    
    followers_of_page_whom_i_follow = followers_this_page_set.intersection(my_following_set)
    
    don_t_followback_me = followers_of_page_whom_i_follow.difference(myfollowers_set)

    driver.back()
    
    return don_t_followback_me
    
```

```python

def interact_with_dialog_and_extract():
    import time 
    
    limit=(500//12)
    
    for countdown in range(limit):
        try:
            element = driver.find_element_by_xpath("//div[@class='oMwYe']")
            driver.execute_script("return arguments[0].scrollIntoView();", element)
    #         print(countdown)
            time.sleep(3)
        except:
            break
        
    # for extracting text from html
    individual_follower=driver.find_element_by_xpath("//div[@class='PZuss']").text


    

                     ############################################
#############################################################################################
#                           DATA CLEANIN STARTS HERE

    #     creating list
    follower_username=individual_follower.split("\nFollow")
    
    
    
    
    only_usernames=[]
    i_j=0
    for fff in follower_username:
        if fff=="" or fff=="ing":  #for tackling empty strings created by split
            continue
        lp=fff.split("\n")

        if i_j==0:   # for tacking first element as it will be split into 2 sep. elements 
                            #while others will be split into 3 elements
            only_usernames.append(lp[0])
            i_j+=1
        else:
            if "Requested" in lp: #for handelling usernames with follow request sent
                only_usernames.append(lp[1])
                
                
                #for handeling continous streak of requested profiles 
                #they are not splited as we used "\nFollow" as a seperator in split
                for count_continous_requested in range(len(lp)):
                    if lp[count_continous_requested]=="Requested" and count_continous_requested!=len(lp)-1:
                        only_usernames.append(lp[count_continous_requested+1])


            else:
                only_usernames.append(lp[1])
                

    return only_usernames
    
```
